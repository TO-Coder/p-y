```python
from pyecharts.charts import Bar
from pyecharts import options as opts
```

```python
init_opts = opts.InitOpts(
    width='1200px',
    height='600px',
    bg_color='black',
    page_title='2403班可视化成绩',
    theme='white'
)
bar = Bar(init_opts)
bar.add_xaxis(['A','B','c'])
bar.add_yaxis('分数',['95','96','78'])

bar.set_global_opts(
    title_opts=opts.TitleOpts(
        title = '2403班学生成绩',
        subtitle = '3位学生成绩',
        pos_left='center',
        title_link = 'http://www.baidu.com',
        title_textstyle_opts=opts.TextStyleOpts(color='red')
    ),
    xaxis_opts=opts.AxisOpts(
        name = '学生姓名'
     ),
     yaxis_opts=opts.AxisOpts(
        name='分数'
    ),
    legend_opts=opts.LegendOpts(
        pos_left='left'
    )
)
bar.set_series_opts(
    label_opts=opts.LabelOpts(
        position="top",
        color='red',
        font_size=15-
    ),
    itemstyle_opts=opts.ItemStyleOpts(
        color="green",
        border_color='black',
        border_width=3,
        border_radius=10
    )
    
)
bar.render('./tmp/bar_2403.html')
```
