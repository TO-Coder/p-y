```python
import pandas as pd
import numpy as np
from pyecharts.charts import Bar
from pyecharts import options as opts
```

```python
df = pd.read_excel("py_echarts_data/商家A和商家B的各类商品的销售数据.xlsx", index_col=0)
df
```

```python
A = df.loc['商家A'].tolist()
B = df.iloc[1].tolist()
tit = df.columns.tolist()
```

```python
init_opts = opts.InitOpts(
    width='1000px',
    height='600px',
    theme='whilt'
)
bar = Bar(init_opts)
bar.add_xaxis(tit)
bar.add_yaxis('商家A',A)
bar.add_yaxis('商家B',B)

# 标题样式
bar.set_global_opts(
    title_opts=opts.TitleOpts(
        title='商家A和商家B销售情况柱状图',
        pos_left='center',
    ),
    xaxis_opts = opts.AxisOpts(
        name = '类型'
    ),
    
    yaxis_opts = opts.AxisOpts(
        name = '销量'
    ),
    # 标题位置
    legend_opts=opts.LegendOpts(
        pos_left='left'
    ),
    # 工具箱
    toolbox_opts= opts.ToolboxOpts(
        is_show=True
    )
)

# 柱子样式
bar.set_series_opts(
    label_opts=opts.LabelOpts(
        position='top',
        color='red',
        font_size=18
    ),
    itemstyle_opts=opts.ItemStyleOpts(
        color="red"
    ),
)

bar.render('./tmp/bar_商家销售情况.html')
```
