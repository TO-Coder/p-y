```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Pie
```

```python
df = pd.read_excel('某淘宝店铺的订单转化率统计数据.xlsx')
stage_names = df.iloc[:, 0].tolist()
user_counts = df.iloc[:, 1].tolist()
```

```python
df
```

```python
data_pairs = []
for i in range(len(stage_names)):
    data_pairs.append((stage_names[i], user_counts[i]))
```

```python
pie = Pie()
pie.add(
    '人数',data_pairs,
    rosetype="area",
    label_opts=opts.LabelOpts(
        formatter="{b}: {c}人 ({d}%)",
        font_size=12,
        position="outside"
    )
    
)

pie.set_global_opts(
    title_opts=opts.TitleOpts(
        title='某淘宝店铺订单分析',
        pos_right='center',
        pos_top='top'
    ),
    legend_opts=opts.LegendOpts(
        pos_bottom='bottom',
        pos_left='center',
        orient='horizontal'
    )
)

pie.set_series_opts(
        label_opts=opts.LabelOpts(
        position='outside',  # 标签显示在外部
        formatter="{b}: {d}%"
    )
)

pie.render('店铺订单转化率分析.html')
```
