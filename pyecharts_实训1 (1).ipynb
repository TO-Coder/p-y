```python
import pandas as pd
from pyecharts.charts import Bar
from pyecharts import options as opts
```

```python
app_categories = ["社交媒体", "视频娱乐", "游戏", "学习阅读", "⽣活⼯具", "其他"]
weekly_hours = [15, 12, 8, 10, 5, 3]
```

```python
init_opts = opts.InitOpts(
    width='900px',
    height='500px',
    theme='dark'
)

bar = Bar(init_opts)

bar.add_xaxis(app_categories)
bar.add_yaxis('小时',weekly_hours)

bar.set_global_opts(
    title_opts=opts.TitleOpts(
        title='学⽣⼀周 App 使⽤时⻓',
        subtitle='数据来源：模拟调查',
        pos_left= 'center'
    ),
    legend_opts=opts.LegendOpts(
        is_show=False
    ),
    xaxis_opts=opts.AxisOpts(
        name = 'App 类别'
    ),
    yaxis_opts=opts.AxisOpts(
        name = '时⻓（⼩时）'
    ),
        tooltip_opts=opts.TooltipOpts(
        trigger="axis",                     # 触发方式：坐标轴触发
        formatter="{b}<br/>时⻓：{c} ⼩时",   # 显示类别和时长
        axis_pointer_type="shadow"          # 指示器样式：阴影
    ),
    toolbox_opts=opts.ToolboxOpts(
        is_show=True
    )
)

bar.set_series_opts(
    label_opts=opts.LabelOpts(
        position='insideTop'
    ),
    itemstyle_opts=opts.ItemStyleOpts(
    )
)
bar.render('./tmp/实训1.html')
```
