```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Pie
```

```python
df = pd.read_excel('商家A和商家B的各类商品的销售数据.xlsx',index_col=0)
```

```python
df
```

```python
col = df.columns.tolist()
sj_A = df.iloc[0].tolist()
sj_B = df.iloc[1].tolist()
```

```python
A = []
B = []
for i in range(len(sj_A)):
    A.append((col[i],sj_A[i]))
    B.append((col[i],sj_B[i]))
```

```python
pie_A = Pie()
pie_A.add(
    '商家A', A,
    label_opts=opts.LabelOpts(
        formatter="{b}:{d}%"
    )
)
pie_A.set_global_opts(
    title_opts=opts.TitleOpts(
        title='商家A各类商品销售额占比',
        pos_left='center'
    ),
    legend_opts=opts.LegendOpts(
        pos_top='bottom'
    )
)
pie_A.render('商家A各类商品销售额占比.html')
```

```python
pie_B = Pie()
pie_B.add(
    '商家B', B,
    label_opts=opts.LabelOpts(
        formatter="{b}:{d}%"
    )
)
pie_B.set_global_opts(
    title_opts=opts.TitleOpts(
        title='商家B各类商品销售额占比',
        pos_left='center'
    ),
    legend_opts=opts.LegendOpts(
        pos_top='bottom'
    )
)
pie_B.render('商家B各类商品销售额占比.html')
```
